```python
from tradingtools.data import dataloader_factory, DataLoaderConfig, DataLoaderTypes
from tradingtools.utils import float_to_decimal
```

```python
pairs = ["A/B", "C/D"]

dummy_data_path = Path.cwd().parent / "tests/data/dummy_data"
list(dummy_data_path.glob("*"))
```

```python
config = DataLoaderConfig(
    type=DataLoaderTypes.historical,
    pairs=pairs,
    interval="1M",
    burn_in_interval=0,
    update_tol_interval="1M",
    hist__parent_path=dummy_data_path,
    hist__sleep_interval="0M",
)

dl = dataloader_factory(config)
```

```python
for data in dl.load():
    print(data.get_latest())
```

```python
df_latest = (data._df.tail(data._n_rows_update_tol)
    .groupby("symbol", maintain_order=True)
    .agg(pl.col("close").drop_nulls().last())
)
```
